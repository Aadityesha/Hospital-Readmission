```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the dataset
df = pd.read_csv('diabetic_data.csv')

# List of continuous and categorical variables for reference
continuous_vars = ['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications',
                   'number_outpatient', 'number_emergency', 'number_inpatient', 'number_diagnoses']

categorical_vars = ['race', 'gender', 'age', 'weight', 'admission_type_id', 'discharge_disposition_id', 
                    'admission_source_id', 'max_glu_serum', 'A1Cresult', 'change', 'diabetesMed', 'readmitted']

# 1. Histogram to check distribution and skewness
def plot_histograms(df, variables):
    for var in variables:
        plt.figure(figsize=(10, 6))
        sns.histplot(df[var], kde=True)
        plt.title(f'Distribution of {var}')
        plt.show()

# 2. Boxplot to detect outliers
def plot_boxplots(df, variables):
    for var in variables:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=df[var])
        plt.title(f'Boxplot of {var}')
        plt.show()

# 3. Bar plot for categorical variables
def plot_barplots(df, variables):
    for var in variables:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=var, data=df)
        plt.title(f'Bar Plot of {var}')
        plt.show()

# 4. QQ Plot to check normality
def plot_qqplot(df, variables):
    for var in variables:
        plt.figure(figsize=(10, 6))
        stats.probplot(df[var], dist="norm", plot=plt)
        plt.title(f'QQ Plot of {var}')
        plt.show()

# 5. Pairplot for continuous variables
def plot_pairplot(df, variables):
    sns.pairplot(df[variables])
    plt.show()

# Visualizations
plot_histograms(df, continuous_vars)
plot_boxplots(df, continuous_vars)
plot_barplots(df, categorical_vars)
plot_qqplot(df, continuous_vars)
plot_pairplot(df, continuous_vars)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from scipy import stats

df = pd.read_csv('diabetic_data.csv')

# 1. Bar Plots for Categorical Variables
categorical_vars = ['race', 'gender', 'admission_type_id', 'discharge_disposition_id', 'admission_source_id']

for var in categorical_vars:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=var)
    plt.title(f'Distribution of {var}')
    plt.xticks(rotation=45)
    plt.show()

# 2. Heatmap of Missing Values
plt.figure(figsize=(10, 6))
msno.heatmap(df)
plt.title('Missing Data Heatmap')
plt.show()

# 3. Histograms for Integer Variables
integer_vars = ['num_lab_procedures', 'time_in_hospital', 'num_procedures', 'num_medications', 
                'number_outpatient', 'number_emergency', 'number_inpatient', 'number_diagnoses']

for var in integer_vars:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[var], bins=20, kde=True)
    plt.title(f'Histogram of {var}')
    plt.show()

# 4. Box Plots for Outlier Detection
for var in integer_vars:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, y=var)
    plt.title(f'Boxplot of {var}')
    plt.show()

# 5. QQ Plots for Normality Check
for var in integer_vars:
    plt.figure(figsize=(6, 6))
    stats.probplot(df[var].dropna(), dist="norm", plot=plt)
    plt.title(f'QQ Plot of {var}')
    plt.show()

# 6. Bar Plot for Target Variable ('readmitted')
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='readmitted')
plt.title('Distribution of Readmission Status')
plt.show()

# 7. Missingness Matrix
plt.figure(figsize=(10, 6))
msno.matrix(df)
plt.title('Missing Data Matrix')
plt.show()

```
